```python
import datetime
from datetime import datetime

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import pickle
import os
import sys
sys.path.append(os.path.join(os.environ['PWD'],'scripts'))

from scrap import get_comments
from preprocess_text import preprocess_raw_comments

from transformers import AutoTokenizer, AutoModel
import torch

from scipy.cluster import  hierarchy
from sklearn.metrics.pairwise import cosine_distances
from sklearn.metrics import silhouette_score

import matplotlib.image as mpimg
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from gensim.models.phrases import Phrases, ENGLISH_CONNECTOR_WORDS
import spacy
```

## Collect all comemnts

```python
data_path = os.path.join(os.environ['PWD'],'data/all_comments.pkl')

if not os.path.isfile(data_path):
    start_date = datetime.strptime('2000-01-01',"%Y-%m-%d")
    end_date = datetime.strptime('2023-01-01',"%Y-%m-%d")
    
    raw_comments_all = get_comments(start_date, end_date) 
    
    pickling_on = open(data_path,"wb")
    pickle.dump(raw_comments_all, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    raw_comments_all = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

```python
len(raw_comments_all)
```

## Preprocess textual data for analysis

```python
data_path = os.path.join(os.environ['PWD'],'data/all_comments_processed.pkl')

if not os.path.isfile(data_path):
    pos_comments = []
    neg_comments = []
    january_index = []
    for comment in raw_comments_all:
        pos_comments.append(comment[0][0])
        neg_comments.append(comment[1][0])
        if comment[2] >= datetime.strptime('2022-01-01',"%Y-%m-%d") and comment[2] < datetime.strptime('2022-02-01',"%Y-%m-%d"):
            january_index.append(True)
        else:
            january_index.append(False)
    
    preprocessed_text_all = {}
    preprocessed_text_all['pos_comments'] = pos_comments
    preprocessed_text_all['neg_comments'] = neg_comments
    preprocessed_text_all['january_index'] = january_index
    
    pickling_on = open(data_path,"wb")
    pickle.dump(preprocessed_text_all, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    preprocessed_text_all = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

## Analyse comments

### Load models for numerical representation of text

```python
model_path = os.path.join(os.environ['PWD'],'HF_models/paraphrase-multilingual-MiniLM-L12-v2')

if not os.path.exists(model_path):

    tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
    model = AutoModel.from_pretrained('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2') 
    
    tokenizer.save_pretrained(model_path)
    model.save_pretrained(model_path)
else:
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModel.from_pretrained(model_path)
    print('models loaded successfully!')
```

### Get numerical representation of text

```python
data_path = os.path.join(os.environ['PWD'],'data/all_pos_embeddings.pkl')

if not os.path.exists(data_path):

    comments = preprocessed_text_all['pos_comments']
    comments_batch = []
    for index in range(len(comments)//64+1):
        comments_batch.append(comments[index*64:min((index+1)*64,len(comments))]) 
    
    encoded_input = []
    # Tokenize sentences
    for comments in comments_batch:
        encoded_input.append(tokenizer(comments, padding=True, truncation=True, return_tensors='pt'))

    output_embeddings = []
    # Compute token embeddings
    with torch.no_grad():
        for embeddings in encoded_input:
            output_embeddings.append(model(**embeddings)['pooler_output'].numpy())

    all_pos_embeddings = np.concatenate(output_embeddings)
    
    pickling_on = open(data_path,"wb")
    pickle.dump(all_pos_embeddings, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    all_pos_embeddings = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

```python
all_pos_embeddings.shape
```

```python
data_path = os.path.join(os.environ['PWD'],'data/all_neg_embeddings.pkl')

if not os.path.exists(data_path):

    comments = preprocessed_text_all['neg_comments']
    comments_batch = []
    for index in range(len(comments)//64+1):
        comments_batch.append(comments[index*64:min((index+1)*64,len(comments))]) 
    
    encoded_input = []
    # Tokenize sentences
    for comments in comments_batch:
        encoded_input.append(tokenizer(comments, padding=True, truncation=True, return_tensors='pt'))

    output_embeddings = []
    # Compute token embeddings
    with torch.no_grad():
        for embeddings in encoded_input:
            output_embeddings.append(model(**embeddings)['pooler_output'].numpy())

    all_neg_embeddings = np.concatenate(output_embeddings)
    
    pickling_on = open(data_path,"wb")
    pickle.dump(all_neg_embeddings, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    all_neg_embeddings = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

```python
all_neg_embeddings.shape
```

### Create linkage values for different hierarchical models

```python
data_path = os.path.join(os.environ['PWD'],'data/all_linkage.pkl')

if not os.path.exists(data_path):
    
    methods = ['single', 'complete', 'average', 'weighted']
    all_linkage = {}
    data = np.concatenate((all_pos_embeddings, all_neg_embeddings))
    
    for m in methods:
        Z = hierarchy.linkage(data, m, metric="cosine")
        all_linkage[m] = Z
        
    pickling_on = open(data_path,"wb")
    pickle.dump(all_linkage, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    all_linkage = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

```python
all_linkage.keys()
```

### Get Silhouette scores

```python
data_path = os.path.join(os.environ['PWD'],'data/all_silhouette_score.pkl')

if not os.path.exists(data_path):
    
    all_linkage_ss = {}
    distance = cosine_distances(np.concatenate((all_pos_embeddings, all_neg_embeddings)))
    
    for link in all_linkage.keys():
        ss = [0,0]
        for i in range(2,401):
            labels = hierarchy.fcluster(all_linkage[link], i, criterion="maxclust")
            ss.append(silhouette_score(distance, labels, metric='precomputed'))
        all_linkage_ss[link] = ss
        
    pickling_on = open(data_path,"wb")
    pickle.dump(all_linkage_ss, pickling_on)
    pickling_on.close()
    print('data pickled successfully!')
else:
    pickle_off = open(data_path, 'rb')
    all_linkage_ss = pickle.load(pickle_off)
    print('data unpickled successfully!')
```

### Word clouds for topics

```python
comments = preprocessed_text_all['pos_comments'] + preprocessed_text_all['neg_comments']
```

```python
len(comments)
```

```python
# Load nlp pipeline
nlp = spacy.load('en_core_web_sm')
# Disable NER component
nlp.disable_pipe('ner')
# Add some stop words to improve analysis results
nlp.Defaults.stop_words |= {'amazon','work','job','good'}
```

```python
processed_text = []
for comment in comments:
        comment = nlp(comment)
        text = []
        for word in comment:
            # Exclude special characters, stopwords, punctuations and numbers
            if len(word.text.replace(" ", "")) != 0 and word.text.replace(" ", "")[0] != "\r" and not word.is_stop\
            and not word.is_punct and not word.like_num:
                text.append(word.lemma_.lower().replace(" ", ""))
        
        if len(text) > 0:
            processed_text.append(text)
        else:
            processed_text.append(' ')
            
# Create bigrams, dictionary and corpus
pos_bigram = Phrases(processed_text, min_count=1, threshold=2, connector_words=ENGLISH_CONNECTOR_WORDS)
processed_text = [pos_bigram[line] for line in processed_text]
processed_text = [' '.join(line) for line in processed_text]
```

```python
len(labels)
```

```python
cluster_wordcloud = {}
for cluster in np.unique(labels):
    text = [processed_text[i] for i in range(len(labels)) if labels[i]==cluster]
    text = ' '.join(text)

    wc = WordCloud(max_font_size=50, max_words=20, background_color="white").generate(text)
    
    cluster_wordcloud[cluster] = wc
```

```python
plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
